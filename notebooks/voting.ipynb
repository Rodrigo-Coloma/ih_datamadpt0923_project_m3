```python
import pandas as pd
import numpy as np
import pickle
from sklearn.ensemble import VotingRegressor
from sklearn.model_selection import cross_val_score


```

```python
df_train = pd.read_csv('./data/clean/diamondsdl123fexytdad2xso37.csv',)
df_test = pd.read_csv('./data/clean/testdl123fexydatdso3.7x2.csv',)

y = df_train['price'].values[1:]
X = np.array(df_train.drop(['Unnamed: 0','price'], axis=1))[1:]
X_test = np.array(df_test.drop('Unnamed: 0',axis=1))
```

```python
xgb = pickle.load(open('./models/xgartdl123dlgridtdso3.7x2_511.sav', 'rb'))
lgb = pickle.load(open('./models/lgdl123dlgridtdso3.7x2_511.sav', 'rb'))
```

```python
vot = VotingRegressor(estimators=[('xgb', xgb),('lgb',lgb)],weights=[6,1])
```

```python
cross_val_score(vot,X,y,scoring='neg_root_mean_squared_error').mean()
```

```python
vot.fit(X,y)
```

```python
preds = vot.predict(X_test)
output = pd.DataFrame({'id': list(range(len(preds))),
                       'price': preds.squeeze()})
```
